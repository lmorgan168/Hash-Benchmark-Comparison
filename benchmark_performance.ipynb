```python
#Step 1: Import data

import pandas as pd
import matplotlib.pyplot as plt


filepath = "D:\\UCI MDS Winter Quarter 2022\\CS 261P\\Project Folder\\project2"

data = pd.read_csv(filepath+"\\Dictionary Data Structure times.csv", index_col = 0)
```

```python
#Step 2
print(data)
```

```python
data = data.replace('NULL', None)
data = data.apply(pd.to_numeric, errors = 'coerce')
```

```python
print(data)
```

```python
#Step 3:
plt.figure(figsize=(10,6))
```

```python
n_elements = data.columns.astype(int)
structures = data.index
```

```python
for structure in structures:
    plt.plot(n_elements, data.loc[structure], marker = 'o', label=structure)
plt.title('Benchmark Performance of Dictionary Structures')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Elapsed Time (s)')
plt.legend(title='Structure')
plt.grid()

outputfile = "benchmark_performance.png"
plt.savefig(outputfile, dpi=300, bbox_inches='tight')
plt.show()
```
